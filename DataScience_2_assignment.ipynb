8.performing similar analysis on another data set tips which is pre-loaded in seaborn library

```python
#Loading a data called tips
tips=sns.load_dataset('tips')
tips.head()
```

```python
fig,ax=plt.subplots(figsize=(8,5))
total_bill_sex = tips.groupby('sex')['total_bill'].mean()
total_avg_bill =round(total_bill_sex,2)
num_quarters=len(total_avg_bill)
colormap=plt.cm.get_cmap('viridis',num_quarters)
total_avg_bill.plot(kind='bar',color=[colormap(i) for i in np.arange(num_quarters)],ax=ax)
ax.set(ylabel='Averge_total_bill',title='Averege_bill_per_sex')
```

```python
#Detecting outliers per tip
fig,ax=plt.subplots(figsize=(8,5))
plt.boxplot(tips['tip'],patch_artist=True,notch=True)
ax.set(xlabel="Distibution of tip",ylabel='tip_values',title='Detection of outliers for tip')
plt.show()
```

```python
color_palette = sns.color_palette('coolwarm')
plt.figure(figsize=(8,6))
ax = sns.histplot(x='total_bill',data=tips, palette=color_palette,hue='day')
for p in ax.patches:
    ax.annotate(format(p.get_height()), (p.get_x() + p.get_width() / 2, p.get_height()),
               ha="center", va ="center", xytext = (0,5), textcoords="offset points")
```

Handling null values using thyroidDF dataset from Kaggle

```python
thyroid =pd.read_csv('/content/thyroidDF.csv')
thyroid.head()
```

```python
thyroid.isnull().sum()
```

```python
thresh =(len(thyroid)//100)*70
```

```python
thresh
```

```python
useless_columns=[]
for col in thyroid.columns:
  if thyroid.isnull().sum()[col]>thresh:
    useless_columns.append(col)
```

```python
useless_columns
```

```python
#since 'TBG' contains more than 70 % of null values we can drop the column
thyroid.drop('TBG',axis=1,inplace=True)
```

```python
thyroid.info()
```

```python
#handling the null vales for the categorical column sex
thyroid['sex'].isnull().sum()
```

```python
thyroid['sex'].iloc[1]
```

```python
thyroid['sex'].iloc[-1]
```

```python

#since first and last values are not null we can use ffill and bfill methods for handling the null values
#using ffill values to fill the null values
ffill=thyroid['sex'].fillna(method='ffill')
```

```python
type(ffill)
```

```python
values=ffill.value_counts()
color=['pink','blue']
num_values=thyroid['sex'].nunique()
```

```python
ax =values.plot(kind='bar',color=[color[i] for i in np.arange(num_values)])

plt.ylabel('count')
for p in ax.patches:
    ax.annotate(format(p.get_height()), (p.get_x() + p.get_width() / 2, p.get_height()),
               ha="center", va ="center", xytext = (0,5), textcoords="offset points")
```

```python
#using back fill method to handle null values
bfill=thyroid['sex'].fillna(method='bfill')
values_bfill=bfill.value_counts()
```

```python
ax =values_bfill.plot(kind='bar',color=[color[i] for i in np.arange(num_values)])

plt.ylabel('count')
for p in ax.patches:
    ax.annotate(format(p.get_height()), (p.get_x() + p.get_width() / 2, p.get_height()),
               ha="center", va ="center", xytext = (0,5), textcoords="offset points")
```

```python
thyroid['T3'].isnull().sum()
```

```python
thyroid['T3'].std()
```

```python
mean_method=thyroid["T3"].fillna(thyroid["T3"].mean())
print(f'Standard deviation of the T3 column replaced by the mean is {mean_method.std()}')
```

```python
median_method=thyroid["T3"].fillna(thyroid["T3"].median())
print(f'Standard deviation of the T3 column replaced by the median is {median_method.std()}')
```

```python
mode_method=thyroid["T3"].fillna(thyroid["T3"].mode()[0])
print(f'Standard deviation of the T3 column replaced by the mode is {mode_method.std()}')
```

```python
interpolate_method=thyroid["T3"].interpolate()
print(f'Standard deviation of the T3 column replaced by the interpolation is {interpolate_method.std()}')
```

```python
#methods like ffill,bfill etc
ffill_method=thyroid["T3"].fillna(method='ffill')
print(f'Standard deviation of the T3 column replaced by the forwardfill is {ffill_method.std()}')
```

```python
bfill_method=thyroid["T3"].fillna(method='bfill')
print(f'Standard deviation of the T3 column replaced by the backwardfill is {bfill_method.std()}')
```
